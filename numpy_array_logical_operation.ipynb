# Logical Operations with Numpy Arrays

```python
import numpy as np
```

```python
array_1 = np.array([3,5,7,2])
array_2 = np.array([[4,1,6,9],
                    [0,5,7,3]])
```

You can compare multiple arrays and perform logical operations. 

```python
array_1 > array_2  #The result is a boolean array of the same shape as the highest-shaped array in the expression
```

You can compare array to a number

```python
array_1 > 2
```

```python
3 < array_1
```

You can execute equality logical expressions

```python
array_1 == array_2
```
