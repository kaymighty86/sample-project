# Importing and Exporting Numpy Arrays
> To or from either npy file or csv file

```python
import numpy as np
```

```python
numpy_arr = np.random.randint(5, 30, size=(4,5)) #generate a random 2d array
```

### For .npy file

Exporting in .npy file (numpy file)

```python
np.save("exported_numpy_array.npy", numpy_arr) #numbers exported are in floating point numbers
```

importing from .npy file (numpy file)

```python
array_frm_npy = np.load("exported_numpy_array.npy")
print(array_frm_npy)
```

### For .csv file

Exporting to .csv file

```python
np.savetxt("exported_numpy_array.csv", numpy_arr, delimiter=",")#numbers exported are in floating point numbers
```

Importing array from .csv file

```python
array_frm_csv = np.loadtxt("exported_numpy_array.csv", delimiter=",")
print(array_frm_csv)
```
