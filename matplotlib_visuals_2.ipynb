# Cutomisation of Matplotlib Visuals
You can fully customise your chart in anyway    

### Full Chart with Info

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
courses = np.array(["Maths", "Data Sci", "Astro Phy", "Comp Sci", "Nursing"])
fees = np.array([fee for fee in np.random.randint(20,40, (5,))])
colors = ["red", "blue", "green", "yellow", "#42c2f5"]

#plot the bar chart with all the necessary customisations
plt.bar(courses, fees, color=colors)
plt.title("Average Fees per Course", fontdict={"size": 20})
plt.xlabel("Courses")
plt.ylabel("Fees ($)")

#customise the range ticks on the y axis
y_range = [num for num in range(0, 51, 5)]
plt.yticks(y_range, [f"${num}k" for num in y_range])

#show the chart
plt.show()
```

### Chart Combo

```python
students_no = [num * 10 for num in np.random.randint(2,5,5)]

plt.bar(courses, fees, color=colors)
#add line graph of students population per course
plt.plot(courses, students_no, lw = 2, c="orange", label = "Student Number")
plt.title("Average Fees per Course", fontdict={"size": 20})
plt.xlabel("Courses")
plt.ylabel("Fees ($)")

#customise the range ticks on the y axis
y_range = [num for num in range(0, 51, 5)]
plt.yticks(y_range, [f"${num}k" for num in y_range])

plt.legend() #show the legend

plt.show()
```

Combo Example 2

```python
series_1 = np.random.randint(20, 100, 12)
series_2 = np.random.randint(5, 50, 12)
series_3 = np.random.randint(40, 70, 12)

months = np.array(["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"])

plt.plot(months, series_1, label="Wind Speed 1")
plt.plot(months, series_2, label="Wind Speed 2")
plt.plot(months, series_3, label="Wind Speed 3")

plt.legend(loc="upper right")
plt.show()
```

### Sub Plots

> Plotting multiple charts in seperated figures at once

```python
seasons = ["winter", "spring", "summer", "autumn"]
student_admitted = [num * 1000 for num in np.random.random(4)]

plt.figure(1)
plt.bar(seasons, student_admitted)

plt.figure(2)
plt.pie(student_admitted, labels=seasons)

plt.show()
```

> Plotting multiple charts in one figure.  
> I'm also covering how to export visuals to PNG image

```python
fig,axs = plt.subplots(2,2) #i'm using a 2D shape (1,2) for the figure but it can also be 1D (e.g. (1,2))

# print(axs.shape)
axs[0, 0].plot(seasons, student_admitted)
axs[0, 1].pie(student_admitted, labels=seasons)
axs[1, 0].bar(seasons, student_admitted)
axs[1, 1].scatter(seasons, student_admitted)

fig.suptitle("Analysing Student Admissions per Season")

plt.savefig("Student Admissions.png") ##export the visual as PNG image

plt.show()
```
