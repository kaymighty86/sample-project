# 3D Visuals in Matplotlib

> 3D visuals are not intteractable in Jupyter. You need to run the code on a normal Python file to enable Matplotlib load the window that allows interraction with visuals

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_series = np.random.randint(10,100, 100)
y_series = np.random.randint(10,100, 100)
z_series = np.random.randint(10,100, 100)
# y_series = np.array([np.sin(num) for num in x_series])
# z_series = np.array([np.cos(num) for num in x_series])

threeDPlot = plt.axes(projection="3d")
threeDPlot.scatter(x_series, y_series, z_series)
plt.show()
```

Example with line graph

```python

x_series = np.arange(0,50, 0.5)
y_series = np.array([np.sin(num) for num in x_series])
z_series = np.array([np.cos(num) for num in x_series])

threeDPlot = plt.axes(projection="3d")
threeDPlot.plot(x_series, y_series, z_series)
plt.show()
```

### 3D Mesh Visual

```python
x = np.arange(-5,5, 0.1)
y = np.arange(-5,5, 0.1)

X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

threeDPlot = plt.axes(projection="3d")

threeDPlot.plot_surface(X,Y,Z, cmap="Spectral") #the "cmap" parameter is "color mapping". Its what gives th visual the color distribution (by default surface visual shows as blue)

plt.show()
```
