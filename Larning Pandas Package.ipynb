# Learning Pandas Package
## Pandas is used fo all things datasets

```python
import pandas as pd
```

**create a panda series using list (column of data)**

```python
eye_color = pd.Series(["red","brown","blue","green","purple"]) #this is a Series object
ages = pd.Series([45, 26, 14,45,12]) #this is another series object

print(f"Eye Color Series: \n{eye_color}") #show series
print()
print(f"Ages Series: \n{ages}") #show series
```

**Create a Data Frame**

```python
eye_data = pd.DataFrame({
    "eye_color": eye_color, 
    "ages": ages
})

eye_data #in Jupyter, you can also print an object by just typing the name
```
