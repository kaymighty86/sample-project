# Performing Statistics with Numpy Arrays 

```python
import numpy as np
```

```python
nd_arr_0 = np.array([[2,3,4,5],
                   [9,8,7,6],
                   [3,4,0,1]])
```

**SUM**

```python
sum_of_all = np.sum(nd_arr_0)
print(sum_of_all)
```

**AVERAGE**

```python
avg = np.mean(nd_arr_0)
print(avg)
```

**MEDIAN**

```python
median = np.median(nd_arr_0)
print(median)
```

**MAX & MIN**

```python
max = np.max(nd_arr_0)
min = np.min(nd_arr_0)
print(max)
print(min)
```

**STANDARD DEVIATION**

```python
std = np.std(nd_arr_0)
print(std)
```
