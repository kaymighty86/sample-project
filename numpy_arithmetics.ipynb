Arrays can be used in arithmentic calculations since they are matrices

```python
import numpy as np
```

```python
array1 = np.array([23,4,6])
array2 = np.array([9,8,3])
```

```python
sumMatrix = array1 + array2 #simple arithmetics can be done with array (if you do this with default python list-type objects, it will just concatenate it)
print(sumMatrix)
```

```python
print(array1 / array2) #divide
```

```python
print(array1 * array2) #multiply
```

**You can perform arithmetics on arrays of different shapes**
> BUT one of the arrays must be a simple shape of either single row or single column

```python
array3 = np.array([[8,4,5],
                   [23,9,2]])
array4 = np.array([[3],
                   [4]]) #single column array

print(array3 + array4)
```

```python
array5 = np.array([[8,4,5],
                   [23,9,2]])
array6 = np.array([3,9,1]) #single row array

print(array5 + array6)
```
