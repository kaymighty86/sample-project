# SERIES DATATYPE

The "Series" datatype is a custom 1D data structure from the Pandas package similar to the default 1D data structures in Python but with more robust functions and behaviours for working with data

```python
import pandas as pd
```

Series are created/casted from a list-type data structure (e.g. lists, tupples, sets)

```python
names_series = pd.Series(["James","Husker","Mary","Florence"])
names_series
```

**Series data strucures can be initialised with custom indexes**

>This replaces the default indexes generated by Series during initialisation

```python
names_series2 = pd.Series([120,44,52,5,77,95], index=["Ade","Pe",2,1,4,3]) #you can assign the same indexes to multiple elements
names_series2
```

**Getting values from Series**

>Series can be sliced just like lists, tupples, etc

```python
print(names_series[0])
print(names_series[2])
```

```python
names_series[::2] #can slide with start, stop, step
```

**Another way of getting data from series**

```python
#getting data using the location identifier: the assigned indexes (custom indexes)
print(names_series2.loc[4]) #this will return all elements with the same indexes (NOTE!! make sure the index exists)
#OR
#getting value using the position index (the index of the location of the element "iloc". Does the same thing as just slicing the default Python way)
print(names_series2.iloc[1])
```

> You can slice with start, stop, step using "loc" and "iloc" method too

```python
names_series2.loc[:3] #remember this uses the assigned indexes not the positional index
```

```python
names_series2.iloc[::2]
```
