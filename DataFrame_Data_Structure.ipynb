# Dataframe Datatype

The "Dataframe" datatype is a custom 2D data structure from the Pandas package similar to the default dictionary in Python but its more **tabular** in structure and with more robust functions and behaviours for working with structured data

```python
import pandas as pd
```

Dataframes are created by casting Dictionary of Series, Lists, tupples, or sets into Dataframes

```python
sales_dataframe = pd.DataFrame({
    "produt_name": ["Fan", "Box", "Plate", "Car"],
    "price": [180,300,130,550]
})

sales_dataframe
```

**Slicing Column**  
Slice with the key (The way you'll normally do with dicts

```python
make = car_sales["Make"] #this outputs the Series stored in the "Make" column
print(make, end="\n\n")
#OR
#you can get the slice by calling as an object (like javascriopt objects)
print(car_sales.Make)
```

**Slicing Row**

> You can slice out a row using its **assigned index** (remember this can be custom index list assigned during the creation of the dataframe (Just like how you do with Series))

```python
sale_record_1 = car_sales.loc[0] #this is slicing a row using its assigned index
sale_record_1 #this outputs a Series containing the values from row 1 and the headers as the assigned indexes (remeber, custom indexes can also be string not only int)
```

```python
type(sale_record_1) #confirm the type of the output (its Series)
```

> **You can slice out a row using its _positional index_ (using iloc[id])**

```python
sales_record_2 = car_sales.iloc[0]
sales_record_2
```

> **You can slice a DataFrame out with either "loc" or "iloc"**

```python
sales_record_3 = car_sales.loc[2:5]
sales_record_3
```

### Filtering a DataFrame

You can filter out a DataFrame based on a condition with any of the values

```python
toyotaSales = car_sales[car_sales["Make"] == "Toyota"]
print(f"Toyota Sales:\n{toyotaSales}", end="\n\n")

odometerAbove5k = car_sales[car_sales["Odometer (KM)"] > 60000]
print(f"Odometer above 60k:\n{odometerAbove5k}")
```

```python
toyotaSales.mean(numeric_only=True) #show the average of all the numerical values in the Dataframe. "numeric_only" bool attribute must be provided to tell Python that its only the numerical value it should perform the operation on. So that an error will not be returned because of other data types in the dataframe
```
