```python
import matplotlib.pyplot as plt
import numpy as np
```

### Scatter Plot

```python
#lets generate random data and visualise how house size affect house price with scatter plot
house_size = [int(size * 1000) for size in np.random.random(200)]
house_price = [int(size * 30000) for size in np.random.random(200)]

plt.scatter(house_size, house_price, c="green", s=20, marker="x", alpha=0.5) #params are x-axis, y-axis, ...others (c => color, s => scale, alpha => alpha, marker => "o")
plt.show()
```

### Line Graph

```python
#lets generate random data and visualise how house price change over time with line graph
year = [year + 2010 for year in range(10)]
price = [int(size * 30000) for size in np.random.random(10)]

plt.plot(year, price, c="red", lw=2)
plt.show()
```

### Bar Chart

```python
#using the data above lets show the scale of this time series data with bar chart
plt.bar(year, price, color="red")
plt.show()
```

### Histogram

```python
#Histogram is not the same as bar chart, as its used for visualising normal distribution
ages = np.random.normal(20, 1.5, 200) #random module has a function for generating a random array with the normal distributon. The params are, the mean, standard devistion, number of elements 

plt.hist(ages)
plt.show()
```

For histograms, you can specify the number of Bins to show in the chart

```python
plt.hist(ages, bins=5)
plt.show()
```

On histograms, you can plot cumulative histograms

```python
plt.hist(ages, bins = 20, cumulative = True)
plt.show()
```

From the visual above, we can see that there are more people above age of 20

### PIE CHART

```python
#lets look at the distribution of people's prog lang preferences
prog_lang = ["Python", "Java", "C++", "C#", "Go"]
votes = [votes * 100 for votes in np.random.random(5)]
explode_slice = [0,0,0.2,0,0] #we're telling matplotlib to visually slice out the portion of C++ by 5 units distance

plt.pie(votes, labels=prog_lang, explode=explode_slice, autopct="%.2f%%") #autopct => auto percent (to show the percent values on the pie in 2 decimal place, the double percent "%%" at the back is just to show the percent sign (two is used becuase one "%" sign is already a reserved character and it means modulus in Python))
plt.show()
```

### Box Plots
Box plots are used for visualising the median and the other percentile scales of data

```python
#Lets generate a randome normal distribution of ages with mean of 45, std of 10, and 100 elements
ages_normal_dist = np.random.normal(45, 10, 100)

plt.boxplot(ages_normal_dist)
plt.show()
```

Box plots show the min, 1st quartile, median (2nd quartile), 3rd quartile, max
